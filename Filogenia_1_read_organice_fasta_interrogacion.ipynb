```python
# Este script Python contiene una función llamada verificar_patron que utiliza expresiones regulares para verificar si una cadena cumple con un patrón
# específico. Luego, la función itera sobre una lista de cadenas, verifica cada cadena utilizando la función verificar_patron y
# agrega las cadenas que cumplen el patrón a una lista llamada accesions. Finalmente, imprime la lista accesions y su longitud.

import re
def verificar_patron(cadena):
    # Expresión regular para el patrón de dos letras mayúsculas seguidas de seis dígitos
    patron = r'^[A-Z]{2}\d{6}$' ############ cambiar esto si es otro patron
    if re.match(patron, cadena):
        return True
    else:
        return False

accesions = [ ]
# Iterar sobre la lista de strings y verificar el patrón
for cadena in lista:
    if verificar_patron(cadena):
        accesions.append(cadena)

print(accesions)
len(accesions)
```

**7. Crear una carpeta donde guardar todas los accesos. Tras crearla, modifica el path en el filename sin borrar lo de "{accession}.gb"

```python
# Se utiliza Entrez.efetch para recuperar los registros de secuencias nucleotídicas correspondientes a las accesiones en la lista 'accesions'
handle = Entrez.efetch(db='nucleotide', id=accesions, rettype='gb', retmode='text')
# Se utiliza SeqIO.parse para analizar los registros recuperados en formato GenBank
registros = list(SeqIO.parse(handle, 'genbank'))
handle.close()  # Se cierra el handle después de recuperar los registros

# Se itera sobre cada registro recuperado
for record in registros:
    accession = record.id  # Se obtiene el ID de la accession del registro
    # Se construye el nombre del archivo utilizando el ID de la accession y se especifica la ubicación donde se guardará el archivo
    filename = f"C:\\Users\\arramirezj\\Desktop\\prueba\\{accession}.gb" ########### path
    # Se abre el archivo en modo de escritura y se guarda el registro en formato GenBank utilizando SeqIO.write
    with open(filename, "w") as output_file:
        SeqIO.write(record, output_file, "genbank")
    # Se imprime un mensaje indicando que el registro se ha guardado correctamente en el archivo especificado
    print(f"Registro {accession} guardado en {filename}")
```

```python
import os
from Bio import SeqIO

# Ruta de la carpeta que contiene los archivos GenBank
carpeta_origen = r"C:\Users\arramirezj\Desktop\prueba"


# Función para obtener el género de una secuencia a partir del archivo GenBank
def obtener_genero(archivo): # ruta
    with open(archivo, 'r') as handle: # abre archivo y guarda como handle. es lo mismo que escribir handle =  open(archivo, "r")
        for record in SeqIO.parse(handle, 'genbank'): #  itero por cada secuencia en el archivo handle con formato genbank
            for feature in record.features: #iterando dentro de cada una de las propiedades del registro del gb (locu, etc) ,  me meto en cada propiedad, atributo (variable que tiene la clase (seqRecords)
                if feature.type == 'source' and 'organism' in feature.qualifiers: # si la propiedad es de tipo source y el cualifiers es organismo:
                    return feature.qualifiers['organism'][0].split()[0]  # el género es la primera palabra en el nombre científico,
    return None

# Iterar sobre los archivos GenBank en la carpeta
for archivo in os.listdir(carpeta_origen):
    if archivo.endswith('.gb'):
        ruta_completa = os.path.join(carpeta_origen, archivo)# une el path de la carpeta con el nombre del archivo, para leerlo  para que la funcion sepa que archivo leer
        genero = obtener_genero(ruta_completa) # a partir de la ruta accede al archivo y busca el genero
        print(genero)


        # Si se pudo obtener el género
        if genero: # si existe
            # Crear la carpeta del género si no existe
            carpeta_destino = os.path.join(carpeta_origen, genero) #ruta nueva a la carpeta del genero
            if not os.path.exists(carpeta_destino): # si no existe me la creas
                os.makedirs(carpeta_destino)

            # Mover el archivo a la carpeta del género
            nueva_ruta = os.path.join(carpeta_destino, archivo)
            os.rename(ruta_completa, nueva_ruta) # igual que mover de carpeta
            print(f"Archivo {archivo} movido a la carpeta {genero}")
        else:
            print(f"No se pudo obtener el género para el archivo {archivo}")
```

```python

import os
from Bio import SeqIO

############### cambiar
REGIONS = ["matk", "internal transcribed spacer", "rbcl", "trnl-trnf", "rps8", "18s", "matr"] # forskaolea
# REGIONS = ["ndhf","rbcl", "trnl","trna-leu"] # trnl-trnf intergenic spacer # pancriatumm
def extract_definition(genbank_file, output_dir):
    with open(genbank_file, "r") as file:
        record = SeqIO.read(file, "genbank")
        return record.description.lower(), output_dir

def obtener_archivos_en_subcarpetas(ruta):
    lista_archivos = []
    for carpeta_actual, subcarpetas, archivos in os.walk(ruta):
        for archivo in archivos:
            ruta_archivo = os.path.join(carpeta_actual, archivo)
            lista_archivos.append(ruta_archivo)
    return lista_archivos

# Ruta del directorio de salida ################### cambiar
output_directory = "C:\\Users\\arramirezj\\Desktop\\marcadores_foska"

# Ruta del directorio de entrada ################## cambiar
input_directory = "C:\\Users\\arramirezj\\Desktop\\prueba"

for archivo in obtener_archivos_en_subcarpetas(input_directory):
    definition, _ = extract_definition(archivo, output_directory)
    for region in REGIONS:
        if region in definition:
            # Verifica si "18s" está presente y si "internal transcribed spacer" está presente antes de "18s"
            if "18s" in definition and "internal transcribed spacer" in definition.split("18s")[0]: # porque pueden estar los dos, si esta los dos va a ints y si solo 18 s pues a ese
                selected_region = "internal transcribed spacer"  # Selecciona "internal transcribed spacer"
            else:
                selected_region = region
            with open(os.path.join(output_directory, f"{selected_region}.fasta"), "a+") as handle:
                with open(archivo, "r") as file:
                    record = SeqIO.read(file, "genbank")
                    for feature in record.features:
                        if feature.type == 'source' and 'organism' in feature.qualifiers:
                            organism = feature.qualifiers['organism'][0]
                    handle.write(f">{str(record.id)}|{organism}\n")
                    handle.write(f"{str(record.seq)}\n")
            break  # Una vez que se encuentra el marcador, no es necesario buscar otros
```

```python
### aaaa pasar de 60 x linea a todo en una linea y asi el script siguiente puede funcioonar bien
# Nombre del archivo de entrada y salida
archivo_entrada = "C:\\Users\\arramirezj\\Desktop\\its_completa.txt"
archivo_salida = "output.fasta"

# Abrir archivo de entrada y crear archivo de salida
with open(archivo_entrada, "r") as f_in, open(archivo_salida, "w") as f_out:
    # Variable para almacenar el identificador actual
    identificador = None

    # Iterar sobre cada línea del archivo de entrada
    for linea in f_in:
        # Eliminar los espacios en blanco al inicio y al final de la línea
        linea = linea.strip()

        # Verificar si la línea comienza con '>', indicando un identificador
        if linea.startswith(">"):
            # Si es un identificador, escribirlo directamente en el archivo de salida
            if identificador is not None:
                f_out.write("\n")  # Añadir un salto de línea antes de escribir el siguiente identificador
            f_out.write(linea + "\n")
            identificador = linea
        else:
            # Si es una línea de secuencia, escribir la secuencia en una sola línea en el archivo de salida
            f_out.write(linea)

# Informar que el archivo se ha procesado exitosamente
print("Se ha procesado el archivo con éxito.")

#### añade ?? emedio si estan x lineas, si es una unica linea no
# Nombre del archivo de entrada y salida
archivo_entrada = "output.fasta"
archivo_salida = "ouput2.fasta"

# Función para realizar las modificaciones en la secuencia
def modificar_secuencia(secuencia):
    # Verificar si la secuencia no está vacía y contiene caracteres distintos de '-' o '\n'
    if secuencia.strip('-\n'):
        # Modificación desde el principio hasta la primera letra
        indice_inicio = next((i for i, char in enumerate(secuencia) if char not in ['-', '\n']), None)
        secuencia_modificada = '?' * indice_inicio + secuencia[indice_inicio:]

        # Modificación desde la última letra hasta el final
        secuencia_invertida = secuencia_modificada[::-1]
        indice_fin = next((i for i, char in enumerate(secuencia_invertida) if char not in ['-', '\n']), None)
        secuencia_modificada_final = secuencia_invertida[:indice_fin].replace('-', '?') + secuencia_invertida[indice_fin:]
        secuencia_modificada_final = secuencia_modificada_final[::-1]

        return secuencia_modificada_final
    else:
        # Si la secuencia está vacía o solo contiene guiones, devolvemos la secuencia de '?' con la misma longitud
        return '?' * len(secuencia)

# Abrir archivo de entrada y crear archivo de salida
with open(archivo_entrada, "r") as f_in, open(archivo_salida, "w") as f_out:
    for linea in f_in:
        # Modificar la secuencia y escribirla en el archivo de salida
        secuencia_modificada = modificar_secuencia(linea.strip())
        f_out.write(secuencia_modificada + "\n")

print("Se han generado las secuencias modificadas en", archivo_salida)

# Nombre del archivo de entrada y salida
archivo_entrada = "C:\\Users\\arramirezj\\Desktop\\ouput2.fasta"
archivo_salida = "modifier_ALIG_matk_completa+RUTH.fas"

# Función para dividir la secuencia en líneas de 60 caracteres
def dividir_secuencia(secuencia):
    # Dividir la secuencia en líneas de máximo 60 caracteres
    secuencia_dividida = '\n'.join(secuencia[i:i+60] for i in range(0, len(secuencia), 60))
    return secuencia_dividida

# Abrir archivo de entrada y crear archivo de salida
with open(archivo_entrada, "r") as f_in, open(archivo_salida, "w") as f_out:
    for linea in f_in:
        # Verificar si la línea comienza con '>', indicando un identificador
        if linea.startswith(">"):
            # Si es un identificador, escribirlo directamente en el archivo de salida
            f_out.write(linea)
        else:
            # Si es una línea de secuencia, dividir la secuencia en líneas de máximo 60 caracteres y escribirla en el archivo de salida
            secuencia_dividida = dividir_secuencia(linea.strip())
            f_out.write(secuencia_dividida + "\n")

print("Se ha dividido la secuencia en líneas de 60 caracteres y se ha guardado en", archivo_salida)
```
